```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
red_wine_url = "https://archive.ics.uci.edu/ml/machine-learning-databases/wine-quality/winequality-red.csv"
white_wine_url = "https://archive.ics.uci.edu/ml/machine-learning-databases/wine-quality/winequality-white.csv"
```

```python
red_wine = pd.read_csv(red_wine_url, sep=';')
white_wine = pd.read_csv(white_wine_url, sep=';')
```

```python
print("Red Wine Dataset:")
print(red_wine.head())
print("\nWhite Wine Dataset:")
print(white_wine.head())
```

```python
red_wine['type'] = 'red'
white_wine['type'] = 'white'
wine = pd.concat([red_wine, white_wine], ignore_index=True)
```

```python
wine
```

```python
# Check for missing values
missing_values = wine.isnull().sum()
print("\nMissing Values:\n", missing_values)
```

```python
# Summary statistics
print("\nSummary Statistics:\n", wine.describe(include='all'))
```

```python
# Exploratory Data Analysis (EDA)

# Distribution of quality ratings
plt.figure(figsize=(10, 5))
sns.countplot(x='quality', data=wine, palette='deep')
plt.title('Distribution of Wine Quality Ratings')
plt.xlabel('Quality')
plt.ylabel('Count')
plt.show()
```

```python
# Distribution of quality ratings by wine type
plt.figure(figsize=(10, 5))
sns.countplot(x='quality', hue='type', data=wine, palette='viridis')
plt.title('Distribution of Wine Quality Ratings by Type')
plt.xlabel('Quality')
plt.ylabel('Count')
plt.legend(title='Wine Type')
plt.show()
```

```python
# Scatter plot of alcohol vs. quality
plt.figure(figsize=(10, 5))
sns.scatterplot(x='alcohol', y='quality', hue='type', data=wine, palette='pastel')
plt.title('Alcohol Content vs. Quality')
plt.xlabel('Alcohol Content')
plt.ylabel('Quality')
plt.legend(title='Wine Type')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.lineplot(x='quality', y='pH', data=wine, hue='type', palette='bright')
plt.title('pH level vs. Quality')
plt.xlabel('Quality')
plt.ylabel('pH level')
plt.legend(title='Wine Type')
plt.show()
```
